5. Not all the data entries are readable.  Be sure to handle that using try and except.    

````python
# to read and write csv files
import csv

# create a function to read the csv and create a list of its contents
def read_csv():
    # create a blank list
    athlete_list = []
    
    # open .csv file as read only 
    with open('athlete_events.csv', 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            # append athlete list as with row data
            athlete_list.append(row)
            
        return athlete_list

# create a function to get the value for the column you want, i.e. age, ht, wt
def stat_list(x):
    # grab list from csv
    athlete_list = read_csv()
    # intialize blank list for stats
    res = []

    # we start at 1 bc we dont want to include the headers
    for i in range(1,len(athlete_list)):
        
        try:
            # to find the weight, height, and age, change the second index value, x
            val = (athlete_list[i][x])
            # append to the list as an integer
            res.append(int(val))

        # any NA value will be replaced with a 0
        except ValueError:
            val = 0
            res.append(val)
            
            
    return res, athlete_list
    
    

def find_mean(type, x):
    # call function to get result list and athlete list
    res, athlete_list = stat_list(x)
    # mean is sum of all 'n' / count of n        
    mean = sum(res) / len(athlete_list)
    # print the mean, \u2022 is unicode for a bullet point
    print(f'\u2022The mean {type} is {mean}')

def find_median(type, x):
    # call function to get result list and athlete list
    res, athlete_list = stat_list(x)       
    # sort the list in order to find the median        
    res.sort()  

    # if the length of the list is odd...
    if len(res) % 2 != 0:
        # find the middle index
        middle = len(res) // 2
        # determine value of middle index
        median = res[middle]
        
    # else if the length of the list is even..    
    else:
        # find middle value
        middle = len(res) // 2
        # find value below and above middle value
        n = res[middle-1:middle+1]
        # find mean of middle values to find median
        median = (n[0] + n[1]) / 2
        
    # print median, \u2022 is unicode for a bullet point
    print(f'\u2022The median {type} is {median}')


def find_mode(type, x):
    # call function to get results and athlete list
    res, athlete_list = stat_list(x)
    # create a dictionary to count frequency of stats
    count = {}
    
    for n in res:
        # add n to the key of map and set its initial value to 1; +1 each occurence
        count[n] = 1 + count.get(n, 0)
    # find the max value   
    max_number = max(count.values())

    # iterate dictionary, key added to modes when value equal to the max_number
    modes = [k for k, v in  count.items() if v == max_number]

    # the map() fucntion applies the str() function to each item in the modes list
    # so that it can be printed with the join function if there are multiple modes
    print(f'\u2022The mode for {type} is {", ".join(map(str, modes))}\n')

# formatting and printing results
print(f'{'Age':^12}\n {"-" * 24}')
age = find_mean('age', 3), find_median('age', 3), find_mode('age', 3)
print(f'{'Height':^12}\n {"-" * 24}')
height = find_mean('height', 4), find_median('height', 4), find_mode('height', 4)
print(f'{'Weight':^12}\n {"-" * 24}')
weight = find_mean('weight', 5), find_median('weight', 5), find_mode('weight', 5)
````
