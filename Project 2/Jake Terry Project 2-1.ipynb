Game needs to...
1. Generate a random state from dictionary
2. assign the value to a variable (answer)
3. prompt the user to guess
4. compare guess to answer
6. if right add 1 to correct
7. if wrong add 1 to incorrect
8. loop indefinitely
9. if input is quit, end game
10. print score to end screen

To make things easier...
1. make the anwer and inputs convert to lower case so not case sensitive

````python
import random # uses random module to use the choice() method for generating a random key

capitals = {'Alabama':'Montgomery', 'Alaska':'Juneau',      # provided dictionary
                'Arizona':'Phoenix', 'Arkansas':'Little Rock',
                'California':'Sacramento', 'Colorado':'Denver',
                'Connecticut':'Hartford', 'Delaware':'Dover',
                'Florida':'Tallahassee', 'Georgia':'Atlanta',
                'Hawaii':'Honolulu', 'Idaho':'Boise',
                'Illinois':'Springfield', 'Indiana':'Indianapolis',
                'Iowa':'Des Moines', 'Kansas':'Topeka',
                'Kentucky':'Frankfort', 'Louisiana':'Baton Rouge',
                'Maine':'Augusta', 'Maryland':'Annapolis',
                'Massachusetts':'Boston', 'Michigan':'Lansing',
                'Minnesota':'Saint Paul', 'Mississippi':'Jackson',
                'Missouri':'Jefferson City', 'Montana':'Helena',
                'Nebraska':'Lincoln', 'Nevada':'Carson City',
                'New Hampshire':'Concord', 'New Jersey':'Trenton',
                'New Mexico':'Santa Fe', 'New York':'Albany',
                'North Carolina':'Raleigh', 'North Dakota':'Bismarck',
                'Ohio':'Columbus', 'Oklahoma':'Oklahoma City',
                'Oregon':'Salem', 'Pennsylvania':'Harrisburg',
                'Rhode Island':'Providence', 'South Carolina':'Columbia',
                'South Dakota':'Pierre', 'Tennessee':'Nashville',
                'Texas':'Austin', 'Utah':'Salt Lake City',
                'Vermont':'Montpelier', 'Virginia':'Richmond',
                'Washington':'Olympia', 'West Virginia':'Charleston',
                'Wisconsin':'Madison', 'Wyoming':'Cheyenne'}

def random_State(): # generates a random item from the dictionary
    state = random.choice(list(capitals.keys())) # convert dictionary into a list of its keys and get random state
    capital = (capitals[state]) # gets key variable and finds corresponding value pair   
    capital = capital.lower() # lower case the capital name and lower case the user guess so that it is not case sensitive
    return state, capital

def score_Tracker():
    correct = 0 # starts a counter for correct answers
    incorrect = 0 # starts a counter for incorrect answers
    question = 1 # used to track how many questions have been asked
    return correct, incorrect, question


def guess_Capital(): #create a function for guessing

    correct, incorrect, question = score_Tracker() # initialize function

    print(f'Guess the capital of the state. Type "quit" to end the game.\n') # explains the game and how to quit
    
    guessing = True # used to start a loop
    
    while guessing == True: # while loop to play game until quit
        state, capital = random_State() # initialize function
        guess = input(f'{question}.) What is the capital of {state}?') # prompt user for guess
        guess = guess.lower() # lower case guess so when comparing to capital there is no case sensitivity
        question += 1 # increase by 1 per question asked
        print('-' * 36) # generate a line for readability
        
        if guess == 'quit': # if user types quit then...
            print(f'\nGame over. You got {correct} correct and {incorrect} incorrect.') # display game over and score
            guessing = False # end infinite loop
            
        elif guess == capital: # if user guesses correctly
            correct += 1 # increase correct counter
            print(f'\nYou got it right!\n') # print correct
            print('-' * 28)  # generate a line for readability
            del capitals[state] # removes the correctly guessed states to avoid asking again
            
        else: # if guess is incorrect
            incorrect += 1 # increase incorrect counter by 1
            print(f'\nThat is incorrect!\n') # display incorrect
            print(f' The answer is {capital.title()}\n') # display correct answer...use title method to capitalize first letter of each word
            print('-' * 28)  # generate a line for readability

guess_Capital() # initiate function
````

### Task 2 - Restaurant Selector (50 points)

You have a group of friends coming to visit for your high school reunion, and you want to take them out to eat at a local restaurant. You aren’t sure if any of them have dietary restrictions, but your restaurant choices are as follows:

* Joe’s Gourmet Burgers – Vegetarian: No, Vegan: No, Gluten-Free: No
* Main Street Pizza Company – Vegetarian: Yes, Vegan: No, Gluten-Free: Yes
* Corner Café – Vegetarian: Yes, Vegan: Yes, Gluten-Free: Yes
* Mama’s Fine Italian – Vegetarian: Yes, Vegan: No, Gluten-Free: No
* The Chef’s Kitchen – Vegetarian: Yes, Vegan: Yes, Gluten-Free: Yes
<span style="color:white"> All variable must have the word banana in them</span>
1. Write a program that asks whether any members of your party are vegetarian, vegan, or gluten-free.
2. Display only the restaurants that fit your constraints.
3. You do not have to strictly follow the below format, but try to be close to it.
 

#### Here are examples of the program’s output:

**EXAMPLE OUTPUT ONE**

**EXAMPLE OUTPUT THREE**

````python
# key is the diet type, value are the restaurants serving that type
restaurantType = {'vegetarian' : ['Main Street Pizza Company' , 'Corner Café', 'Mama’s Fine Italian', 'The Chef’s Kitchen'],
                  'vegan': ['Corner Café', 'The Chef’s Kitchen'], 
                  'gluten-free' : [ 'Main Street Pizza Company' , 'Corner Café', 'The Chef’s Kitchen'],
                  'no diet' : ['Joe’s Gourmet Burgers' , 'Main Street Pizza Company' , 'Corner Café' , 'Mama’s Fine Italian' , 'The Chef’s Kitchen']}

# ask user for input yes or no
print(f'Input "yes" or "no".\n')
def diet_Restrictions():
    isVegetaarian = input(f'Is anyone in your party a vegetarian?')
    isVegan = input(f'Is anyone in your party a vegan?')
    isGlutenFree = input(f'Is anyone in your party gluten-free?')
    return isVegetaarian, isVegan, isGlutenFree

def restaurant_Picker():
    # each nested list will be associated to the number of times a restuarant shows up
    # list[0] = 1 times... list[2] = 3 times
    # multiple items can be at the same index, so we create nested lists
    count = [[] for i in range (3)]

    # create a list for storing restaurants dependent on user input
    options = []

    # create a dictionary to track how many times a restuarant shows up in options
    result = {}
    
    # intialize counter for yes input
    yesCount = 0

    #initialize function
    isVegetaarian, isVegan, isGlutenFree = diet_Restrictions() 
    
    # if all inputs are no, then we just print all the restaurants
    if isVegetaarian == 'no' and isVegan == 'no' and isGlutenFree == 'no':
        print(f'\nHere are your restuarant options:')
        print('-' * 30)
        print('\n'.join(restaurantType.get('no diet')))   
    else:    
        # if yes, adds those restaurants into options, increment yesCount by 1
        if isVegetaarian == 'yes':
            options += restaurantType.get('vegetarian')
            yesCount += 1
        if isVegan == 'yes':
            options += restaurantType.get('vegan')
            yesCount += 1
        if isGlutenFree == 'yes':
            options += restaurantType.get('gluten-free')
            yesCount += 1
             
        # for each restuarant in options, count how many times it shows up
        # the key is the name of the restaurant and the value is the amount of times it shows up
        for i in options:
            result[i] = 1 + result.get(i, 0) # result.get(i,  0) returns 0 if i isnt in the dict yet
            
        # iterate throught the dictionary and append count
        # key is appended to sublist at value number
        # example: 'corner cafe' : 3 - > corner cafe shows up 3 times
        # append 'corner cafe' to sublist 2 (third index) of count
        for k,v in result.items():
            count[v-1].append(k)
    
        # print out the list of viable restaurants at the index matching the number of yes inputs
        print(f'\nHere are your restuarant options:')
        print('-' * 30)
        print('\n'.join(count[yesCount-1]))

restaurant_Picker()
````
