# Dungeon
# The Rulez:
    # Players take turns moving.
    # Players must roll and move to the end of the board.
    # Each roll has a chance for a random event.
    # Players decide on an action and can result in gaining points, losing points, or nothing.
    # Once all players reach the end of the board, the total points are added up. 
    # Which ever player has the most points wins.

```python
import random # to use random

class Players: #create player class and attributes
    def __init__(self, name): 
        self.name = name # name of player
        self.position = 0 # 0 to start
        self.points = 0 # 0 to start

    def move_player(self, moves): # track players position across the board
        self.position += moves # update the position
        print(f"{self.name} moves to {self.position}.\n") # print new pos
        
    def add_points(self, points):     # method to add points to players attribute
        self.points += points # update pooints
        print(f"{self.name} gains {points} points. Current points are {self.points}.\n") # print new points

    def get_points(self):     # getter for calling points
        return self.points # show points


class Board:     # creat the board, where random event will happen on each space that can add or subtract from player points
    def __init__(self, length): 
        self.length = length # length intializes at start
        # create a list of tuples, with event descriptors and methods
        # create a method for each event
        self.events = [
            ("You find a treasure chest!\n", self.treasure_chest),
            ("A goblin blocks your path!\n", self.goblin_encounter),
            ("You meet a wandering merchant.\n", self.merchant_trade),
            ("You stepped onto a trap!\n", self.hidden_trap)
        ]

    def trigger(self, player):     # function for triggering random events
        event_description, event_logic = random.choice(self.events)      # from the tuples, use random to assign descriptor and logic
        print(f'{player.name}: {event_description}\n')  # print the event
        event_logic(player)         # call the event with the player as an argument

##### Event functions #####
    
    def encounter_mimic(self, player):   # players can encounter a mimic while opening a chest
        print(f'You encounter a mimic!') 
        print(f'1: Fight it (-4 points if you lose, +8 points if you win)') # choice 1
        print(f'2: Run away (-1 point)\n') # choice 2
        choice = self.get_choice([1, 2])  # checks user input
        if choice == 1:
            roll = random.randint(1, 6)
            if roll >= 4:  # Win condition must be a 4 or above
                print("You defeated the mimic!\n")
                player.add_points(8) # update points
            else:
                print("The mimic overpowered you!\n")
                player.add_points(-4) # update points
        elif choice == 2:
            print("You ran away!\n")
            player.add_points(-1) # update points
            

    def treasure_chest(self, player):
        # method for treasure chest event
        print(f'1 - Open the treasure chest (Requires a roll).')
        print(f'2 - Ignore it.\n')
        choice = self.get_choice([1, 2])  # validated user input
        if choice == 2:
            print(f'You ignore the chest.\n')
        # 3 results can happen: open and gain points, fight a mimic, or fail to open
        elif choice == 1:
            roll = random.randint(1, 6)
            if roll == 1:
                self.encounter_mimic(player) # player encounters mimic
            elif roll >= 3:
                player.add_points(5) # update points
                print(f'You found treasure! {player.name} has {player.points}\n') # print new points
            elif roll <3:
                print(f'You failed to open the chest!\n') 

    def goblin_encounter(self, player): # goblin encounter
        print(f"1: Fight it (-2 points if you lose, +4 points if you win)")
        print(f"2: Run away (-1 point)\n")
        choice = self.get_choice([1, 2])  # check user input
        if choice == 1:
            roll = random.randint(1, 6)
            if roll >= 4:  # Win condition
                print("You defeated the goblin!\n")
                player.add_points(4) # update points
            else:
                print("The goblin overpowered you!\n")
                player.add_points(-2) # update points
        elif choice == 2:
            print("You ran away!\n")
            player.add_points(-1) # update points

    def merchant_trade(self, player): # merchant method
        print("1: Trade 3 points for a magic item (+8 points)")
        print("2: Ignore him (no effect)")
        print("3: Rob him! (-8 points if you lose, +12 points if you win)\n")
        choice = self.get_choice([1, 2, 3])  # check user input
        if choice == 1:
            if player.points >= 3: # check if player has enough points
                print("The merchant gives you a powerful magic item!\n")
                player.add_points(8 - 3)  # Net gain of 5 points
            else:
                print("You don't have enough points to trade.\n") # youre broke lol
        elif choice == 2:
            print("You ignored the merchant.\n") # nothing happens
        elif choice == 3:
            roll = random.randint(1, 6)
            if roll >= 5:  # Win condition hard as heck
                print("You robbed the merchant!\n")
                player.add_points(12) # lots of points tho
            else:
                print("The merchant overpowered you!\n")
                player.add_points(-8) # but you lose a ton

    def hidden_trap(self, player): # trap method
        print("1: Try to disarm the trap (risky!)")
        print("2: Walk around it (-1 point)\n")
        choice = self.get_choice([1, 2])  # validated user input
        if choice == 1:
            roll = random.randint(1, 2)
            if roll == 1:
                print("You successfully disarmed the trap!\n") # 50/50 to disarm, nothing happens if you win
            else:
                print("The trap catches you! You lose 3 points!\n")
                player.add_points(-3) # lose points
        elif choice == 2:
            print("You carefully avoid the trap.\n") # you lose 
            player.add_points(-1) # update points

    def get_choice(self, valid_choices):  # checks if an input exists
        while True:
            try:
                choice = int(input(f'Choose: '))
                if choice in valid_choices: # valid_choices defined inside the event methods
                    return choice  # return valid choice
                else:
                    print(f"Invalid choice! Please choose from {valid_choices}.") # wrong chose inputted
            except ValueError:
                print(f"Invalid input! Please enter an integer from {valid_choices}.") # value error if not int


class Game: # class for game attributes
    def __init__(self):
        self.players = [] # list to hold all players initialized
        self.board = Board(24) # initialize Board class at length 24
        self.game_end = False # game end set to false

    
    def setup(self): # function for setting up game
        print(f'This is a dungeon game!\n') # print introduction
    
        # Use a while loop to repeatedly ask for input until it's valid
        while True:
            try:
                num_players = int(input(f'Enter how many will be playing.\n'))
                if num_players >= 2:  # Ensure at least 2 players
                    break  # Exit the loop if the input is valid
                else:
                    print(f"Please enter a valid number of players (minimum 2): \n") # reprompt until correct
            except ValueError:
                print(f"Invalid input! Please enter a valid number.\n") # catch non-int inputs
        
        # Create player objects and add them to the self.players list
        for i in range(num_players): # for each player
            name = input(f'Enter your name player {i + 1}.\n') # ask for a name
            self.players.append(Players(name)) # add to self.players list
            print() # new line
                    
        print(f"The game will begin! \nThe board has {self.board.length} spaces.\nReach the end and gather the most points to win!\n") # start game


    def player_turn(self, player): # function for players taking a turn
        print('-' * 24) # formatting to distinguish turns
        if player.position >= self.board.length: # checks if player is at the end of the board first
            print(f"{player.name} has reached the end!\n") # print if player reached end
        else: # else take player turn
            input(f'{player.name} you have {player.get_points()} points \nPress enter to roll.\n') # player rolls
            roll = random.randint(1, 6)
            print(f"{player.name} rolled a {roll}!\n") # display roll
            player.move_player(roll) # move the player and update pos
        
            # self.board is the instance of the board which contains the events
            self.board.trigger(player)

    def check_game(self): # check game state
        for player in self.players: # check each player
            if player.position < self.board.length: # if one player has not finished, return false and exit function
                return False
        return True # all players have finished will return true

    def winner(self): # function for calculating winner
        print("\nThe game is over! Here are the final results:\n") # print results
        max_points = -9999 # set max points to large negative number guarantee its value will update 
        for player in self.players: # for each player
            print(f"{player.name}: {player.points} points\n") # print out name and total points
            if player.points > max_points: # if players points are greater than max points
                winner = player.name # winner is assigned player name
                max_points = player.points # max points becomes player points, then check the next player
        print(f"\nCongratulations, {winner}! You are the winner with {max_points} points!\n") # player with most points wins

    def play_game(self): # function to setup the game and check if game over
        self.setup() # run setup 
        while not self.check_game(): # while check game not True
            for player in self.players: # check each player
                if self.check_game(): # if check game returns true, break and continue to self.winner
                    break
                self.player_turn(player) # otherwise take the players turn
        self.winner() # announce winner



game = Game() # create a game class
game.play_game() # run the game
```
